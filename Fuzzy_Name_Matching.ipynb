```python
import re
c_df['Clean_name'] = c_df['Name'].apply(lambda x: re.sub(r'"','',x))
```

```python
def fuzz_ratio_function(input_string,database = c_df ):
  c_df['Clean_name'] = c_df['Name'].apply(lambda x: re.sub(r'"','',x))
  best_match_score = 0
  for i in c_df['Clean_name']:
    Ratio = fuzz.ratio(i.lower(),input_string.lower())
    if Ratio>best_match_score:
      best_match_score = Ratio
  return best_match_score

def partial_match_function(input_string,database = c_df ):
  c_df['Clean_name'] = c_df['Name'].apply(lambda x: re.sub(r'"','',x))
  best_match_score = 0
  for i in c_df['Clean_name']:
    Ratio = fuzz.partial_ratio(i.lower(),input_string.lower())
    if Ratio>best_match_score:
      best_match_score = Ratio
  return best_match_score

def token_sort_match_function(input_string,database = c_df ):
  c_df['Clean_name'] = c_df['Name'].apply(lambda x: re.sub(r'"','',x))
  best_match_score = 0
  for i in c_df['Clean_name']:
    Ratio = fuzz.token_sort_ratio(i.lower(),input_string.lower())
    if Ratio>best_match_score:
      best_match_score = Ratio
  return best_match_score

def token_set_match_function(input_string,database = c_df ):
  c_df['Clean_name'] = c_df['Name'].apply(lambda x: re.sub(r'"','',x))
  best_match_score = 0
  for i in c_df['Clean_name']:
    Ratio = fuzz.token_set_ratio(i.lower(),input_string.lower())
    if Ratio>best_match_score:
      best_match_score = Ratio
  return best_match_score





```

```python
t_df['fuzz_ratio'] = t_df['Test Name'].apply(fuzz_ratio_function)
t_df['partial_ratio'] = t_df['Test Name'].apply(partial_match_function)
t_df['token_sort_ratio'] = t_df['Test Name'].apply(token_sort_match_function)
t_df['token_set_ratio'] = t_df['Test Name'].apply(token_set_match_function)

```
